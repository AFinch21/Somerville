```python
data_cut = data.copy()
data_cut = data_cut[['id', 'filename','pre_text', 'post_text', 'table_ori', 'qa']]
data_cut = data_cut[data_cut['qa'].notna()]
```

```python
data_cut
```

```python

# Extracting the desired fields
data_cut['company'] = data_cut['filename'].apply(lambda x: x.split('/')[0])
data_cut['year'] = data_cut['filename'].apply(lambda x: x.split('/')[1])

data_cut['question'] = data_cut['qa'].apply(lambda x: x.get('question'))
data_cut['steps'] = data_cut['qa'].apply(lambda x: x.get('steps'))
data_cut['program'] = data_cut['qa'].apply(lambda x: x.get('program'))
data_cut['answer'] = data_cut['qa'].apply(lambda x: x.get('answer'))
data_cut['exe_answer'] = data_cut['qa'].apply(lambda x: x.get('exe_ans'))
data_cut['str_exe_answer'] = data_cut['exe_answer'].apply(lambda x: x if isinstance(x, str) else None)
data_cut['exe_answer'] = data_cut.apply(
    lambda row: data_cut['exe_answer'][row.name] if row['str_exe_answer'] is None else 0.0,
    axis=1
)
```

```python
data_cut.columns
```

```python
data_cut = data_cut.drop('qa', axis=1)

data_cut = data_cut[['id', 'company',
       'year','filename', 'pre_text', 'post_text', 'table_ori',
       'question', 'steps', 'program', 'answer', 'exe_answer', 'str_exe_answer']]

data_cut
```

## Initial Testing


```python
import requests

# Set the URL for the endpoint
url = "http://localhost:8000/answer_question"

def get_response(question):
# Create the data payload
    data = {
        "message": question,
        "status": "string"
    }

    # Make the POST request
    response = requests.post(url, json=data)

    # Check the response
    if response.status_code == 200:
        print("Response:", response.json())
        return response.json()['message']
    else:
        print("Request failed with status code:", response.status_code)
```

```python
get_response("was was the average cost per share of the 2008 settlement of the 2007 forward repo contract?")
```
