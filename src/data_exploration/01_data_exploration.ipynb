```python
data_cut = data.copy()
data_cut = data_cut[['id', 'filename','pre_text', 'post_text', 'table_ori', 'qa']]
data_cut = data_cut[data_cut['qa'].notna()]
```

```python
data_cut
```

```python
# Extracting the desired fields
data_cut['company'] = data_cut['filename'].apply(lambda x: x.split('/')[0])
data_cut['year'] = data_cut['filename'].apply(lambda x: x.split('/')[1])

data_cut['question'] = data_cut['qa'].apply(lambda x: x.get('question'))
data_cut['steps'] = data_cut['qa'].apply(lambda x: x.get('steps'))
data_cut['program'] = data_cut['qa'].apply(lambda x: x.get('program'))
data_cut['answer'] = data_cut['qa'].apply(lambda x: x.get('answer'))
data_cut['exe_answer'] = data_cut['qa'].apply(lambda x: x.get('exe_ans'))
```

```python
data_cut.columns
```

```python
data_cut = data_cut.drop('qa', axis=1)

data_cut = data_cut[['id', 'company',
       'year','filename', 'pre_text', 'post_text', 'table_ori',
       'question', 'steps', 'program', 'answer', 'exe_answer']]

data_cut
```
